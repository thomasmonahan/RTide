## Generating A Synthetic Tidal Signal:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
### Producing a 1-year reference series sampled at 6' intervals
times = pd.date_range(start = '2024-01-01 00:00:00+00:00', periods=24*365, freq="1h",tz='UTC')
time_vals = times.to_julian_date().to_numpy()

### Computing the contribution of each constituent
u_tides = []
v_tides = []

for i,j in enumerate(cinformation['frq']):
    u_tides.append(cinformation['u'][i]*np.cos(24*j*np.array(time_vals) * 2 * np.pi + (cinformation['theta'][i] *np.pi /180)))
    v_tides.append(cinformation['v'][i]*np.cos(24*j*np.array(time_vals) * 2 * np.pi + (cinformation['g'][i] *np.pi /180)))

### Calculating the final tide as the superposition of all constituents for each time instance
u_tide = np.sum(u_tides, axis = 0) + 0.1*np.random.normal(len(u_tides))
v_tide = np.sum(v_tides, axis = 0) + 0.1*np.random.normal(len(v_tides))
```

```python
### We have a tidal signal!
plt.plot(times, u_tide)
plt.plot(times, v_tide)

plt.xlim(times[0], times[-1])
plt.title('Synthetic tide')
plt.xlabel('Time')
plt.ylabel('Sea Level (m)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

## Simple Application of RTide

RTide can be run using three lines of code. By default, RTide takes pandas dataframes with date-time indexes as input and, for currents, measurements provided in two columns labeled 'u' and 'v'. Note the datetime index MUST be UTC aware. 

In addition to the observations dataframe, RTide also requires the latitude, longitude.

To remove the need to recompute the global tide function in between runs, a file name should be specified to save the computed input functions, inputs, and trained model. An Rtide saves folder will automatically be added to the directory to prevent excess clutter.

```python
import rtide
from rtide import RTide
```

In this basic example we add a different quadratic trends to each component. RTide can calculate both trends independently. 

```python
df = pd.DataFrame({'u': u_tide  , 'v': v_tide}, index = times)

model = RTide(df[:7000], lat = 58.6557103353, lon = 3)

inputs = {
          'symmetrical': True,
          'path': 'simple_currents',
          'fixed_location':True,


}
model.Prepare_Inputs(**inputs)
```

### SIREN Architecture
The SIREN architecture outlined in Response-based prediction of tidal currents can easily be utilized by setting architecture = siren. Generally, it is recommended to increase the regularization_strength and lower the learning rate. Good default settings for these are regularization_strength = 0.1, and lr = 1e-4 respectively but care is needed.

```python
train_inputs = {
    'loss' : 'MSE',
    'save_weights' : True,
    'lr':1e-4,
    'standard_epochs': 500,
    'early_stoppage':30,
    'regularization_strength':0.1,
    'architecture': 'siren' ## can use the SIREN architecture outlined in Response-based prediction of tidal currents
}
model.Train(**train_inputs)
```

```python
model.Predict(df[7000:])
```
